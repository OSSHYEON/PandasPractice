### 9. 2021-11-01 00:00:00 ~ 15:00:00 까지 각 에피소드 별 viewcnt 의 증가량을 데이터 프레임으로 만드시오

```python
video.head()
```

```python
start = pd.to_datetime("2021-11-01 00:00:00")
end = pd.to_datetime("2021-11-01 15:00:00")

start, end
```

```python
(video.ct >= start) & (video.ct <= end)
```

```python
target = video[(video.ct>=start)&(video.ct<=end)].reset_index(drop=True)
```

```python
def check(x):
    result = max(x) - min(x)
    return result

answer = target[['videoname', 'viewcnt']].groupby('videoname').agg(check)
answer
```
