### 5. 'HP', 'Attack', 'Defense', 'Sp.Atk', 'Sp.Def', 'Speed' 간 상관 계수 중 <br><br>가장 절대값이 큰 두 변수와 그 결과값을 구하여라

```python
target = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr().abs().unstack().reset_index().rename(columns={0:"corr"})
target
```

```python
answer = result.iloc[0]
answer
```

### 9. 각 세대가 가진 (Type 1, Type 2)의 쌍 중 한 번 씩만 존재하는 쌍을 구하여라

```python
df.head()
```

```python
target = df[['Type 1', 'Type 2']].value_counts()
target
```

```python
target2 = target[target==1]
target2
```

```python
lst = []
for value in target2.reset_index().values:
    t1 = value[0]
    t2 = value[1]
    
    sp = df[( df['Type 1'] == t1 )&( df['Type 2'] == t2 )]
    lst.append(sp)
    
    
result = pd.concat(lst).reset_index(drop=True).Generation.value_counts().sort_index()
result
```
