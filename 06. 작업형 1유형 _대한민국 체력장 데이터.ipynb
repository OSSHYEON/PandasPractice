### 3. 연령대 별 인원수를 구하여라

```python
def age_category(age):
    age = (age//10) * 10
    return age
```

```python
age_category(df['측정나이']).value_counts()
```

### 5. 남성 중 A 등급과 D 등급의 체지방률 평균의 차이(큰 값에서 작은 값의 차)를 구하여라

```python
df.columns
```

```python
df[(df.측정회원성별=='M') & ( df.등급=="A" )].iloc[:, 4].mean()
```

```python
df[( df.측정회원성별 =="M" )&( df.등급 == "D")].iloc[:, 4].mean()
```

### 6. 여성 중 A 등급과 D 등급의 체중 평균의 차이(큰 값에서 작은 값의 차) 를 구하여라

```python
abs(df[( df.측정회원성별 == "F" )&( df.등급 == "A")].iloc[:, 3].mean() - df[( df.측정회원성별 == "F" )&( df.등급 == "D")].iloc[:, 3].mean())
```

### 7. bmi는 자신의 몸무게를 키의 제곱으로 나눈 값이다. <br><br>데이터의 bmi를 구한 다음 새로운 컬럼을 만들고, 남성의 bmi 평균을 구하여라

```python
df.columns
```

```python
df['bmi'] = df['체중 : kg'] / (df['신장 : cm']/100 *  df['신장 : cm']/100)
df.head()
```

```python
df[df['측정회원성별'] == "M"].bmi.mean()
```

---

### 8. bmi보다 체지방율이 높은 사람들의 체중 평균을 구하여라

```python
df[df['bmi']<= df['체지방율 : %']]['체중 : kg'].mean()
```
