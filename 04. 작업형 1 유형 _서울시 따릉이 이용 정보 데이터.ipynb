### 2. 각 요일별 가장 많이 이용한 대여소의 이용 횟수와 대여소 번호를 데이터 프레임으로 출력하라

```python
df.head()
```

```python
df['day_name'] = pd.to_datetime(df['대여일자'])
df
```

```python
df['day_name'] = df['day_name'].dt.day_name()
df
```

```python
result = df.groupby(['day_name', '대여소번호']).size().to_frame('size').sort_values(['day_name', 'size'], ascending=False).reset_index()
result
# count와 size의 차이?
# count : NaN은 제외하고 계산. 함수이다. groupby.count()시 dataframe반환
# size : NaN 포함해서 계산 - 호출된 객체의 크기 또는 길이 반환. 데이터프레임의 속성이다. groupby.size()시 series반환
# to_frame : 주어진 시리즈를 데이터프레임으로 변환한다.
```

```python
answer = result.drop_duplicates('day_name', keep='first').reset_index(drop=True)
answer
```
