### 1. 여름철 6월, 7월, 8월 이화동이 수영동보다 높은 기온을 가진 시간대는 몇 개인가?

```python
df.dtypes
```

```python
df.time = pd.to_datetime(df.time)
df.dtypes
```

```python
summer = df[df.time.dt.month.isin([6, 7, 8])].reset_index(drop=True)
summer
```

```python
answer = summer[summer.이화동기온 > summer.수영동기온]
len(answer)
```
